### Encontrar el KML


```python
def extract_kml_from_kmz(kmz_path, extract_to='temp_kml'):
    # Descomprime el archivo KMZ en el directorio especificado
    with zipfile.ZipFile(kmz_path, 'r') as kmz:
        kmz.extractall(extract_to)
    
    # Busca el archivo KML extraído
    kml_file = None
    for root, dirs, files in os.walk(extract_to):
        for file in files:
            if file.endswith('.kml'):
                kml_file = os.path.join(root, file)
                break
    
    if kml_file:
        print(f"KML file found: {kml_file}")
    else:
        print("No KML file found in the KMZ.")
    
    return kml_file
```

```python
kml_path = extract_kml_from_kmz(kmz_path)
kml_path
```

### SACAR EL XML DEL KML Y exportar a .txt

```python
def save_kml_to_text_file(kml_path, output_txt='kml_content.txt'):
    with open(kml_path, 'r', encoding='utf-8') as file:
        content = file.read()
    
    with open(output_txt, 'w', encoding='utf-8') as output_file:
        output_file.write(content)

    print(f"El contenido del KML se ha guardado en {output_txt}")
```

```python
save_kml_to_text_file(kml_path)
```

```python
def read_coordinates_from_txt(txt_path):
    with open(txt_path, 'r') as file:
        # Leer el contenido del archivo
        content = file.read().strip()
    
    # Dividir el contenido en coordenadas individuales
    coord_strings = content.split(' ')
    
    # Convertir las coordenadas a formato numérico y guardarlas en una lista
    coordinates = []
    for coord in coord_strings:
        lon, lat, _ = coord.split(',')
        coordinates.append([float(lat), float(lon)])
    
    return coordinates
```
